# 팰린드롬 출력

```python
"""
유효한 팰린드롬 출력하기
주어진 문자열이 팰린드롬인지 확인하라. 대소문자 구분 x, 영문자와 숫자만을 고려
"""
```

### My answer

```python
## 문자열을 소문자로 변환하기, 공백과 특수문자 제거
## 문자열을 뒤집기
## ==으로 내용이 같은지 확인

input_sentence = "A man, a plan, a canal: Panama"

input_sentence = input_sentence.lower()

strs = []

for char in input_sentence:
    if char.isalnum():
        strs.append(char)

reversed_strs = strs[::-1]

if strs == reversed_strs:
    print("Input Sentence:", input_sentence)
    print("Palindrome: True")
else:
    print("Input Sentence:", input_sentence)
    print("Palindrome: False")

```

```python
"""
Point:
- isalnum(): 문자열, 영문자, 숫자인지등을 파악하는 함수, 특수문자-공백등은 False 반환
- 리스트 슬라이싱[::-1]: 리스트를 뒤집는 방법

"""
```

# 문자열 뒤집기

```python
"""
문자열을 뒤집하는 함수 작성, 입력은 문자 배열 이고 리턴 없이 리스트 내부를 직접 조작하라
"""
```

```python
input_sentence = ["h","e","l","l","o"]
reversed_sentence = input_sentence[::-1]
print("Input Sentence:", input_sentence)
print("Reversed Sentence:", reversed_sentence)
```

```python
"""
- s = s[::-1] vs s[:] = s[::-1]
전자는 새로운 객체를 새로 할당하는 것이고, 후자는 기존 객체의 값을 직접 변경하는 것에 해당함. 문자열의 경우 s[:]와 같은 슬라이싱 연산이 직접 값을 변경하는 작업에 해당함

- 그리고 자꾸 offset을 기준으로 위치를 점차 줄여나가면서 순회하는 방법이 제안됨
"""
```

```python
def reverseString(s):
    left_offset = 0
    right_offset = len(s) - 1
    for i in range(len(s)//2):
        temp = s[left_offset + i]
        s[left_offset + i] = s[right_offset - i]
        s[right_offset - i] = temp

input_sentence = ["h","e","l","l","o"]
reverseString(input_sentence)
print("Reversed Sentence:", input_sentence)
```

```python
def reverseString_v2(s):
    left_offset = 0
    right_offset = len(s) - 1
    while left_offset < right_offset:
        s[right_offset], s[left_offset] = s[left_offset], s[right_offset] # 우항의 값을 먼저 조사한다음, 한번에 처리
        right_offset -= 1
        left_offset += 1

input_sentence = ["h","e","l","l","o"]
reverseString_v2(input_sentence)
print("Reversed Sentence:", input_sentence)
```

# 로그 파일 재정렬

```python
"""

로그를 재정렬하라. 기준은 다음과 같다.
1. 로그의 가장 앞 부분은 식별자다.
2. 문자로 구성된 로그가 숫자 로그보다 앞에 온다.
3. 식별자는 순서에 영향을 끼치지 않지만, 문자가 동일할 경우 식별자 순으로 한다.
4. 숫자 로그는 입력 순서대로 한다.

"""
```

```python
input_log = ["dig1 8 1 5 1", "let1 art can", "dig2 3 6", "let2 own kit dig", "let3 art zero"]
# desired output: ["let1 art can", "let3 art zero", "let2 own kit dig", "dig1 8 1 5 1", "dig2 3 6"]

# 식별자 - 문자 로그가 숫자 로그보다 먼저
# 문자 로그랑 숫자 로그를 파싱 -> "dig", "let" 패턴
# 문자 로그의 경우, 식별자 순으로, 숫자 로그는 입력 순서대로

dig_log = []
let_log = []
sorted_log = []

for id in input_log:
    # dig, let 로그 식별
    if id.split()[1].isdigit():
        dig_log.append(id)
    else:
        let_log.append(id)
    
let_log.sort(key=lambda x:(x.split()[1:], x.split()[0])) # 문자 로그의 경우 문자 내용을 기준으로 정렬을 수행하고, 내용이 같은 경우에는 식별자의 내용을 후순위로 기준하여 정렬

# 마지막에
sorted_log += let_log
sorted_log += dig_log

print(sorted_log)

# dig가 있다는 점에 주의


```

# 가장 흔한 단어

```python
"""
금지된 단어를 제외한 가장 흔하게 등장하는 단어를 출력하라. 대소문자를 구분하지 않으며, 구두점 또한 무시한다.
"""
```

```python
from collections import Counter
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit."
banned = ["hit"]

# 단어라고 했으니까 split
# 그리고 banned는 counter에서 제외
# 대소문자 구분 x니까 소문자로 정규화
# 구두점은 제외

paragraph = paragraph.lower().replace(".", '').replace(',', '').split(' ') # 단어 단위로 추출
paragraph = [word for word in paragraph if word not in banned] # 제거
cnt = Counter(paragraph) # 개수 세기
print(cnt.most_common()[0][0])
```

```python
"""
- cnt.most_common(): 가장 많이 세어진 튜플을 반환

"""
```

# 그룹 애너그램

```python
"""
문자열 배열을 입력 받아 애너그램 단위로 그룹핑하라
"""

# 단어를 문자열 단위로 분해, 같은 문자열이 있는 것끼리 그룹핑
```

```python
words = ["eat", "tea", "tan", "ate", "nat", "bat"]

# 규칙을 두고 정렬하여 같은지 비교
dct = {}
for word in words:
    key = "".join(sorted(word))
    if key not in dct:
        dct[key] = []
    
    dct[key].append(word)

print(list(dct.values()))
```

# 가장 긴 팰린드롬 부분 문자열

```python
"""
가장 긴 팰린드롬 부분 문자열을 출력하라
"""
```

```python
input_str = "cbbd"

def check_pal(input):
    reversed_input = input.lower()[::-1]
    return reversed_input == input

# 처음에는 문자열 길이만큼 window size를 잡아서 check_pal수행, 이후 window size를 줄여나가면서 offset을 달리하면서 check_pal 수행
found_answer = False
for window_size in range(len(input_str), 0, -1):
    offset = 0
    while(offset+window_size<len(input_str)):
        sliced_sentence = input_str[offset:offset+window_size]
        if check_pal(sliced_sentence):
            print(f"answer:{sliced_sentence}")
            found_answer = True
            break
        offset += 1

    if found_answer:
        break
    
if not found_answer:
    print("Cannot Find Answer")
        
    

    
```
