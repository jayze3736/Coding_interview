# 두 수의 합

```python
"""
덧셈하여 타겟을 만들 수 있는 배열의 두 수 인덱스를 반환하는 함수
"""
```

```python
nums = [2, 7, 11, 15]
target = 9
```

```python
# 첫번째 풀이 1. 전체 조회(브루트 포스 - 무차별 대입 공격) - O(n^2)
answer = []
for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == target:
            answer = [i, j]
    
print(answer)
# 0 - {1,2,3}, 1 - {2,3}, 2 - {3}, 만약 3이면? -> 3 - {3} 이므로 실행 안됨
# 코드 구현 자체는 간단하지만, 무식한 방법
```

```python
# 두번째 풀이 2. 뺄셈을 이용한 타겟 값 조회

# 더하는 것이 기본 구조이기때문에 target 보다 값이 큰 경우에는 제외 -> 사용하지 않는 인덱스를 먼저 필터링
# 그리고 target에서 뺀 값이 있는지 조사 => target - x0 해서 나온 값이 있는가? -> 그것만 조회
# 딕셔너리를 활용 -> 조회라는 개념이 포함되면 딕셔너리를 적극적으로 활용

dic = {}
answer = []
for i, num in enumerate(nums):
    dic[num] = i

# key: 값, value: 인덱스

for num in nums:
    key = target - num
    
    if key in dic: # 조회
        answer = [dic[num], dic[key]]

print(answer)
```

```python
# 두번째 풀이 개선 2. 뺄셈을 이용한 타겟 값 조회

# 하나의 for문으로 통합

dic = {}
answer = []
for i, num in enumerate(nums):
    dic[num] = i
    key = target - num
    
    if key in dic: # 조회
        answer = [dic[num], dic[key]]

print(answer)
```

# 빗물 트래핑

```python
"""
높이를 입력받아 비 온 후 얼마나 많은 물이 쌓일 수 있는지 계산하라
"""
```

```python
height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
# output = 6

# 패턴은 피벗을 중심으로 양측에 단차가 존재해야함 -> 예를 들어서 1, 0, 2 -> 이런 경우는 물이 1칸 쌓일 수 있음 -> 그러나 1, 1, 2 이러면 못 쌓음 -> 중심 피벗으로부터 양 측의 높이는 중심보다 최소한 커야 쌓이고, 
# 최소 높이만큼 물이 쌓임

# 1. pivot을 기준으로 블록에 물이 쌓이는가? 2. 얼만큼 쌓이는가?

# 1을 알아내기 위해서 pivot 기준으로 왼쪽 max 높이와 오른쪽 max 높이를 알아내야함. 왼쪽 max 높이와 오른쪽 max 높이가 모두 pivot보다 크다면 충족
# 2번은 오른쪽 max 높이와 왼쪽 max 높이중에 가장 작은 값을 pivot 높이에서 빼면 됨

volume = 0
left, right = 0, len(height) - 1
left_max, right_max = height[left], height[right]

while left < right: # 투 포인터를 움직이되, 포인터가 교차될때까지 진행
    left_max, right_max = max(height[left], left_max), max(height[right], right_max)

    if left_max <= right_max: # 왼쪽을 기준으로 volume을 카운팅할때, 기본적으로 오른쪽 장벽이 있어야하고, 이 장벽은 왼쪽 max 보다 커야 물이 가두어짐. 따라서 left_max가 right_max 이하일때까지 volume을 측정
        volume += left_max - height[left]
        left += 1
    else: # 높이가 right가 더 작은 경우 -> left는 장벽 역할, right_max와의 단차만큼 volume 측정
        volume += right_max - height[right]
        right -= 1

print(volume)


# 투 포인터에 대한 감각 부족
# 가정 후에 case를 나누어서 문제를 해결하는 테크닉
# 법칙, 가정 등을 잘 생각해서 풀어야함
```

```python
height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]

# 투 포인터를 활용하는 경우 -> 투 포인터는 언제 사용하지? 어떠한 값에 대한 탐색 또는 조회가 필요할 경우

# 언제 물이 쌓이는가?: 현재 위치(pivot)에서 양측의 최대 높이가 자신의 높이보다 클때 -> 양측의 최대 높이를 조사할 필요가 있음
# 얼만큼 물이 쌓이는가?: 양측의 최대 높이중 최소값을 자신의 높이만큼 뺀 단차만큼 쌓임

# 투 포인터 -> 왼쪽과 오른쪽을 각각 고려 -> 왼쪽 시작부터 순회하되 왼쪽 최대 높이만큼 빼도록 설계 -> 오른쪽 최대 높이는 왼쪽 최대 높이보다 크다고 가정해야함
# 반대로 오른쪽인 경우에는 오른쪽 최대 높이만큼 빼도록 설계 -> 왼쪽 최대 높이가 오른쪽 최대높이보다 크도록 가정

# 왼쪽 장벽과 오른쪽 장벽간의 높이 비교를 우선적으로 한다면, 그 다음에 계산 가능한 단차는 사실상 더 작은 장벽 높이에서 빼주면 되기 때문에

left, right = 0, len(height) - 1
left_max, right_max = 0, 0
volume = 0

while left <= right: # 교차할때까지
    
    # max를 갱신하는 과정
    if left_max < height[left]:
        left_max = height[left]

    if right_max < height[right]:
        right_max = height[right]
    
    # left_max, right_max = max(height[left], left_max), max(height[right], right_max)

    if left_max <= right_max: # 오른쪽 최대 높이가 왼쪽 최대 높이보다 큰 경우
        volume += left_max - height[left]
        left += 1
    else:
        volume += right_max - height[right]
        right -= 1

        
print(volume)




```

# 세수의 합

```python
"""
배열을 입력받아 합으로 0을 만들 수 있는 3개의 엘리먼트를 출력하라.
"""
```

```python
nums = [-1, 0, 1, 2, -1, -4]

# 먼저 sorting -> 순서와 숫자 크기간의 연관관계를 줌
nums = sorted(nums)
print(nums)

# 투 포인터를 활용

# 왼쪽 포인터는 음의 값부터 순회, 오른쪽 포인터는 양의 값부터 순회
left, right = 0, len(nums) - 1
answers = []

# left, right 값을 더함 -> 나온 값이 양수? 음수? -> 음수면 오른쪽 포인터부터 임시 포인터가 순회하여 값이 있는지 조사, 양수면 왼쪽 포인터부터

# 만약에 조사한 값이 없다면, 포인터 옮기기. 이때 옮기는 건 이전에 left, right에서 합산한 값이 양수면 0이 되게하려면 양의 값을 줄여야함 -> 따라서 right 포인터를 옮김

# 만약 음이 나왔다면 음의 값의 크기를 줄여야하므로 left 포인터를 옮김

# 또한, 세수의 합의 부호가 바뀌는 경우에는 그 조합이 안된다는 의미라서 옮겨야 함

while left <= right:
    calc = nums[left] + nums[right]
    
    if calc <= 0: # 합산된 결과가 음수인 경우

        temp = right - 1 # 현재 right 이전의 포인터로 옮김
        while nums[temp] >= 0: # 나머지 값은 양수여야함
            calc_f = calc + nums[temp]

            if calc_f == 0:
                entry = [nums[left], nums[right], nums[temp]]
                if not entry in answers:
                    answers.append(entry)
                break
            
            temp -= 1 # 1씩 줄이기
        
        left += 1
    else: # calc > 0 -> 합산된 결과가 양수인 경우
        temp = left + 1

        while nums[temp] < 0: # 나머지 값은 음수여야함
            calc_f = calc + nums[temp]

            if calc_f == 0:
                entry = [nums[left], nums[right], nums[temp]]
                if not entry in answers:
                    answers.append(entry)
                break
            
            temp += 1 # 1씩 올리기
        
        right -= 1
    

print(answers)


```

```python
nums = [-1, 0, 1, 2, -1, -4]
nums = sorted(nums)
print(nums)
results = []
for i in range(len(nums) - 2):
    if i > 0 and nums[i] == nums[i - 1]:
        continue
    left, right = i + 1, len(nums) - 1
    while left < right:
        sum_t = nums[i] + nums[left] + nums[right]
        if sum_t < 0:
            left += 1
        elif sum_t > 0:
            right -= 1
        else:
            results.append((nums[i], nums[left], nums[right]))

            while left < right and nums[left] == nums[left + 1]:
                left += 1
            while left < right and nums[right] == nums[right - 1]:
                right -= 1
            left += 1
            right -= 1
print(results)
```

# 배열 파티션 I

```python
"""n개의 페어를 이용한 min(a, b)의 합으로 만들 수 있는 가장 큰 수를 출력하라"""
```

```python
nums = [1,4,3,2]

# pair는 두개의 element로 구성됨

# 가장 큰 수를 출력하려면 가장 많은 페어를 활용해야함. 단, 중복은 허용 x

# 배열의 크기가 2n일때, pair는 총 n개가 생성됨

# 따라서 전체 pair를 모두 더하려면 아래와 같이 작성

result = sum(sorted(nums)[::2]) # 정렬후에는 값들이 순서대로 위치하고, 짝수번째 값이 항상 작다는 것을 알기에 짝수번째 값들을 더함
print(result)
```

# 자신을 제외한 배열의 곱

```python
"""
배열을 입력받아 output[i]가 자신을 제외한 나머지 모든 요소의 곱셈 결과가 같도록 출력하라.
"""
```

```python
nums = [1,2,3,4]

# nums[0] -> 2x3x4, nums[1] -> 1x3x4, nums[2] -> 1x2x4, nums[3] -> 1x2x3 => [24, 12, 8, 6]이 나오도록, 자기자신의 위치의 값을 제외한 나머지 값들의 곱을 출력하도록

# 단 나눗셈 금지, O(n)선에서 구현

# 자신을 제외한 나머지

# [a,b,c,d]라고 했을때 우리가 원하는건
# [bcd, acd, abd, abc] 임
# 왼쪽부터 1로 초기화해서 먼저 차례로 곱하면 -> [1, a, ab, abc] 그리고 오른쪽에서 1로 초기화해서 차례로 곱하면 -> [bcd,cd,d,1]
# 따라서 이 둘을 곱하면
# [bcd, acd, abd, abc]가 됨

arr = []
temp = 1
for i in range(len(nums)):
    arr.append(temp)
    temp *= nums[i]

temp = 1
for i in range(len(nums)-1, -1, -1): # 0 ~ len(nums) - 1
    arr[i] *= temp
    temp *= nums[i]
    
answer = arr
print(answer)
```

# 주식을 사고팔기 가장 좋은 시점

```python
"""
한 번의 거래로 낼 수 있는 최대 이익을 산출하라
"""
```

```python
import sys
items = [7, 1, 5, 3, 6, 4]
# 1일때 사서 6일때 팔면 차익 5가 생김
# items의 인덱스 순서는 시간 순서를 의미하고, 각 값은 주가를 의미

min_value = sys.maxsize
max_profit = 0

# 배열은 왼쪽부터 순차적으로 조회하기 때문에, 기본적으로 시간 순서 흐름이 깔려있음
for item in items:
    
    if min_value > item:
        min_value = item
    
    profit = item - min_value
    if max_profit < profit:
        max_profit = profit

# min값을 갱신 -> 각 item과의 차익(profit)이 최대가 되는 경우를 구함 -> 기본적으로 조회가 시간 순서로 되어있기때문에 차익을 계산했을때 최대가 되는 경우가 결국에는 최대 수익이 됨

print(max_profit)
    

```
